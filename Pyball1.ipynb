## Let's PlyBall


```python
from pybaseball import schedule_and_record
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## A Quick Analysis of the Dodgers defeated 2017 club vs the 1988 World Series club. 

```python
Dodgers88 = schedule_and_record(1988, 'LAD')
Dodgers17 = schedule_and_record(2017, 'LAD')
```

```python
Dodgers88.describe()
```

```python
Dodgers17.describe()
```

## Let's take a look the win count as the season advances!

```python
Dodgers88['win-count'] = np.where(Dodgers88['W/L']=='W', 1, 0).cumsum()
Dodgers17['win-count'] = np.where(Dodgers17['W/L']=='W', 1, 0).cumsum()
```

```python
plt.plot(Dodgers88['win-count'],label=" ''88 Blue Crew")
plt.plot(Dodgers17['win-count'],label=" '16 Blue Cew")
plt.legend(loc=4)
plt.xlabel('Games into Season')
plt.ylabel('Win Count')
plt.title('Record Throughout Season');
```

## Let's look into the score differnetials? How bad was each team womping on the others 

```python
Dodgers88['scorediff'] = (Dodgers88['R'] - Dodgers88['RA']).cumsum()
Dodgers17['scorediff'] = (Dodgers17['R'] - Dodgers17['RA']).cumsum()

```

```python
plt.plot(Dodgers88['scorediff'],label=" '88 Doyers")
plt.plot(Dodgers17['scorediff'],label=" '17 Doyers")
plt.legend(loc=4)
plt.xlabel('Games into Season')
plt.ylabel('Runs Scored - Runs Against')
plt.title('Cumulative Run Differential');
```
